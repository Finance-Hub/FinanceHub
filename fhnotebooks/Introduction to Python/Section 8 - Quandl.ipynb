## Fetching Data
As an example we will grab the time series of the WTI Crude Oil Price from the *Energy Information Administration* (EIA - US Department of Energy). A quick search on quandl's website lets you find that the ticker is `EIA/PET_RWTC_D`.

```python
WTIdata = quandl.get('EIA/PET_RWTC_D')
WTIdata.plot(figsize=(12, 8))
```

## Searching for Data
Quandl organizes their data based on the source, not on the type. So for example, if your looking for data on **positioning on sugar derivatives** it is easier to find what you are looking for by searching "CFTC" first and than by searching by "sugar". Then you will find the "comitment of traders" report.

```python
COTdata = quandl.get('CFTC/080732_FO_ALL')
```

```python
COTdata.columns
```

```python
COTdata['Open Interest'].plot()
```

## Yield Curve Example
US Zero-coupon nominal curve from the FED.

```python
YCdata = quandl.get('FED/SVENY')
```

```python
YCdata.columns
```

```python
YCdata[['SVENY01', 'SVENY10']].plot()
```

### Caveats
Quandl is great for futures prices from the major exchanges from around the world and great for data from the US. But other than that, there is not much detail. For example you cannot find the series for PETR4 or APPL stocks.
